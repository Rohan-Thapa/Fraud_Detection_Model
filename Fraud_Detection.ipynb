## Now its time for Feature Engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.head()
```

```python
df_model = df.drop(['nameOrig', 'nameDest', 'isFlaggedFraud'], axis=1)
```

```python
df_model.head()
```

```python
categorical = ['type']
numerical = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
```

```python
y = df_model['isFraud']
X = df_model.drop('isFraud', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical),
        ('cat', OneHotEncoder(drop='first'), categorical)
    ],
    remainder='drop'
)
```

```python
pipeline = Pipeline([
    ('prep', preprocessor),
    ('clf', LogisticRegression(class_weight='balanced', max_iter=1000))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(f"The accuracy of the model is {pipeline.score(X_test, y_test) * 100}%") # the percentage of accuracy
```

```python
import joblib

joblib.dump(pipeline, "fraud_detection_pipeline.pkl") # this will be used for prediction system creation.
```
